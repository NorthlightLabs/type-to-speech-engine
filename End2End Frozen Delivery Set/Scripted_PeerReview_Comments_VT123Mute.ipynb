VIDEO TASK 1: VARIABLES AND DATA TYPES

```python
# Lets get started
## Here we will assign the variables.
company_name = "Jackson Coding"
# Stores the company name as a string.
##Stores the company name as a string.
accounting_period = "2022-2023"
# Stores the financial reporting period as a string.
revenue = 1500000.00
# Stores total revenue as a float.
##Stores total revenue as a float.
expenses = 1200000.00
# Stores total expenses as a float.
is_profitable = revenue > expenses
# Returns True if revenue is greater than expenses.
##Returns True if revenue is greater than expenses.

# Now lets code for printing.
## Here I will actually add extra steps to provide a clearer output.
print(f"Company name is: {company_name}", type(company_name))
# Displays the company name and shows that its data type is str.
## The f function to print Company name is was not explicitly asked for in assignment but added for clarity.

print(f"For accounting period: {accounting_period}", type(accounting_period))
# Displays the accounting period and shows that its data type is str.

print(f"Revenue is: £{revenue:,.2f}", type(revenue))
# Displays revenue in currency format and shows that its data type is float.
## Here we have added code to format revenue as currency.

print(f"Expenses are: £{expenses:,.2f}", type(expenses))
# Displays expenses in currency format and shows that its data type is float.

print(f"Is company profitable?: {is_profitable}", type(is_profitable))
# Displays the boolean result and shows that its data type is bool.
## Now that all of the coding is typed, and the comments added, lets go ahead and run the code.
```

VIDEO TASK 2: ARITHMETIC OPERATIONS

```python
# We shall define some calculations.
## Here we will code some basic accounting functions.
gross_profit = revenue - expenses
# Calculates gross profit by subtracting expenses from revenue.

profit_margin = (gross_profit / revenue) * 100
# Calculates profit margin as a percentage of revenue.
##Calculates profit margin as a percentage of revenue.

profit_margin = round(profit_margin, 2)
# Rounds the profit margin to 2 decimal places for clearer presentation.

print(f"Gross Profit: £{gross_profit:,.2f}")
# Displays gross profit in currency format.
## Displays gross profit in currency format.

print(f"Profit Margin: {profit_margin:.2f}%")
# Displays profit margin as a percentage to 2 decimal places.
## Now that this has been coded, lets run this cell.
```

VIDEO TASK 3: CONTROL STRUCTURES-IF STATEMENTS

```python
# Here we will code our if-elif-else formula.
## And here we will code our if elif else formula.
if gross_profit > 0:
    performance = "Profit"
    # Runs when gross profit is greater than zero.

elif gross_profit == 0:
    performance = "Break-even"
    # Runs when revenue exactly equals expenses.
    ## Runs when revenue exactly equals expenses.

else:
    performance = "Loss"
    # Runs when expenses are greater than revenue.

print(f"{company_name} for the period {accounting_period} recorded a {performance}")
# Prints a summary statement using the company name, accounting period, and result.

## This final line of code allows us to print a summary statement which contains the company name, the accounting period, and the result. Lets run this cell and see our result.
```
