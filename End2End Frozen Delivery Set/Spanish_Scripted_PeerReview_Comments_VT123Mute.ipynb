TAREA DE VIDEO 1: VARIABLES Y TIPOS DE DATOS

```python
# Empecemos
## Aquí asignaremos las variables.
company_name = "Jackson Coding"
# Almacena el nombre de la empresa como una cadena.
##Almacena el nombre de la empresa como una cadena.
accounting_period = "2022-2023"
# Almacena el período contable como una cadena.
revenue = 1500000.00
# Almacena los ingresos totales como un float.
##Almacena los ingresos totales como un float.
expenses = 1200000.00
# Almacena los gastos totales como un float.
is_profitable = revenue > expenses
# Devuelve True si los ingresos son mayores que los gastos.
##Devuelve True si los ingresos son mayores que los gastos.

# Ahora vamos a escribir el código para imprimir.
## Aquí añadiré algunos pasos extra para ofrecer una salida más clara.
print(f"Company name is: {company_name}", type(company_name))
# Muestra el nombre de la empresa y enseña que su tipo de dato es str.
## El uso de una f-string para imprimir Company name is no se pidió explícitamente en la tarea, pero se añadió para mayor claridad.

print(f"For accounting period: {accounting_period}", type(accounting_period))
# Muestra el período contable y enseña que su tipo de dato es str.

print(f"Revenue is: £{revenue:,.2f}", type(revenue))
# Muestra los ingresos en formato de moneda y enseña que su tipo de dato es float.
## Aquí hemos añadido código para dar formato de moneda a los ingresos.

print(f"Expenses are: £{expenses:,.2f}", type(expenses))
# Muestra los gastos en formato de moneda y enseña que su tipo de dato es float.

print(f"Is company profitable?: {is_profitable}", type(is_profitable))
# Muestra el resultado booleano y enseña que su tipo de dato es bool.
## Ahora que todo el código está escrito y los comentarios añadidos, vamos a ejecutar el código.
```

TAREA DE VIDEO 2: OPERACIONES ARITMÉTICAS

```python
# Definiremos algunas funciones.
## Aquí programaremos algunas funciones contables básicas.
gross_profit = revenue - expenses
# Calcula el beneficio bruto restando los gastos de los ingresos.

profit_margin = (gross_profit / revenue) * 100
# Calcula el margen de beneficio como un porcentaje de los ingresos.
##Calcula el margen de beneficio como un porcentaje de los ingresos.

profit_margin = round(profit_margin, 2)
# Redondea el margen de beneficio a 2 decimales para una presentación más clara.

print(f"Gross Profit: £{gross_profit:,.2f}")
# Muestra el beneficio bruto en formato de moneda.
## Muestra el beneficio bruto en formato de moneda.

print(f"Profit Margin: {profit_margin:.2f}%")
# Muestra el margen de beneficio como porcentaje con 2 decimales.

## Ahora que esto está programado, vamos a ejecutar esta celda.
```

TAREA DE VIDEO 3: ESTRUCTURAS DE CONTROL - SENTENCIAS IF

```python
# Aquí programaremos nuestra fórmula if-elif-else. ## Y aquí programaremos nuestra fórmula if elif else.
if gross_profit > 0:
    performance = "Profit"      # Se ejecuta cuando el beneficio bruto es mayor que cero.

elif gross_profit == 0:
    performance = "Break-even"  # Se ejecuta cuando los ingresos son exactamente iguales a los gastos. ## Se ejecuta cuando los ingresos son exactamente iguales a los gastos.

else:
    performance = "Loss"        # Se ejecuta cuando los gastos son mayores que los ingresos.

print(f"{company_name} for the period {accounting_period} recorded a {performance}")
# Imprime una frase de resumen utilizando el nombre de la empresa, el período contable y el resultado.

## Esta línea final de código nos permite imprimir una frase de resumen que contiene el nombre de la empresa, el período contable y el resultado. Vamos a ejecutar esta celda y ver el resultado.
```
