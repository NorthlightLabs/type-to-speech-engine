```python
## Lets start by importing the libraries required

# Import libraries
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
## Lets set some parameters
# Set parameters
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2)) * 10 + 50
surf = ax.plot_surface(X, Y, Z, cmap='magma', edgecolor='none')
## Lets name our graph elements
# Naming graph and axes
ax.set_title("Surface Plot: Profit Surface")
ax.set_xlabel("Pricing")
ax.set_ylabel("Sales Volume")
ax.set_zlabel("Profit")
## Some other stuff
# To complete code
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()



```
